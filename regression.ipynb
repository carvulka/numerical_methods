```python
import numpy
import matplotlib.pyplot as pyplot
import pandas
```

```python
# 2.

# 'least sum of squares' method
def polynomial_regression(xs, ys, degree):
    matrix = numpy.array([numpy.array([sum(x ** (j + k) for x in xs) for j in range(degree + 1)]) for k in range(degree + 1)])
    rhs = numpy.array([sum(y * x ** j for x, y in zip(xs, ys)) for j in range(degree + 1)])
    solution = numpy.linalg.solve(matrix, rhs)

    def polynomial(x):
            y = 0
            for i, coeff in enumerate(solution):
                y += coeff * x ** i
            return y
    return polynomial
```

```python
# 1.

xs = numpy.array([0, 1, 2, 3, 4])
ys = numpy.array([0, 1, 1, 2, 2])

x_axis = numpy.linspace(xs.min(), xs.max(), 1000)

pyplot.scatter(xs, ys)
pyplot.plot(x_axis, polynomial_regression(xs, ys, 1)(x_axis), "black")
pyplot.show()
```
